Axes должна принадлежать только одной области Figure. Как правило, всегда сначала создается область Figure, а затем с помощью add_subplot() в Figure размещается одна или несколько областей Axes. 

**Чаще всего изменяем axes! Попробуем изменить и записать несколько параметров.**  
Так удобнее, если не помнишь все параметры, подсказчик подскажет

```python

fig = plt.figure()
ax = fig.add_subplot(111) # We'll explain the "111" later. Basically, 1 row and 1 column.

fig.set_facecolor('green')

ax.set_facecolor('red') # Раскрасим
ax.set_xlim([-10, 10]) # Ось икс от и до
ax.set_ylim([-2, 2]) # Ось игрек от и до
ax.set_title('Название картинки') # Название картинки
ax.set_xlabel('ось абцис (XAxis)') # Название оси абсцисс
ax.set_ylabel('ось ординат (YAxis)') # Название оси ординат

plt.show()
```

**Для рисования на Axes необходимо использовать какой-нибудь из его методов. Методов куча, основные: plot и scatter.  
-plot рисует точки соединенные линиями;  
-scatter просто рисует точки**


```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([0, 1, 2, 3, 4], [0, 6, 7, 15, 19]) # Эти будут соединены линиями
ax.scatter([0, 1, 2, 3, 4], [1, 3, 8, 12, 27]) # Это будут точки

plt.show()
```

```python
# Здесь тоже дофига разных параметров, раскрасим, поменяем размер, например
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([0, 1, 2, 3, 4], [0, 6, 7, 15, 19], linewidth=6, color='red')
ax.scatter([0, 1, 2, 3, 4], [1, 3, 8, 12, 27], marker='*', linewidths=3, color='green')

plt.show()
```

```python
# Тоже самое!!! X и Y пропишем отдельно и вставим, как параметр
x= [0, 1, 2, 3, 4]
y_1 = [0, 6, 7, 15, 19]
y_2 = [1, 3, 8, 12, 27]

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y_1,
        color = 'black',
        linewidth = 5)

ax.scatter(x, y_2,
           color = 'blue',
           marker = '*')

plt.show()
```

**Все методы axes присутствуют в модуле pyplot. Например, при вызове plt.title('spam') модуль pyplot вызывает   ax.set_title('spam'). Можно сказать, что модуль pyplot создает Figure и Axes автоматически (хотя это не совсем   так).**

Так можно и так короче и быстрее, но только на простых графиках и в случае, если надо быстро визуализировать.  
Но на больших проектах так НЕЛЬЗЯ!!! Т.к. все это не очевидно и усложнит чтение!

### Несколько Axes на одном figure

Каждая область Axes является независимой от других, то есть на на них могут быть нарисованы самые разные графики    и установлены самые разные параметры внешнего вида. 

```python
# Рассмотрим пример, как добавить несколько графиков
fig = plt.figure(figsize=(15, 3)) # Просто размер каждого графика, можно менять, как надо по странице

ax_1 = fig.add_subplot(2, 2, 1) # Расположение!!! См начало тетрадки
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)
ax_4 = fig.add_subplot(2, 2, 4)

ax_1.set(title = 'ax_1', xticks=[], yticks=[]) # Соответственно названия каждого и пустая разметка
ax_2.set(title = 'ax_2', xticks=[], yticks=[])
ax_3.set(title = 'ax_3', xticks=[], yticks=[])
ax_4.set(title = 'ax_4', xticks=[], yticks=[])

plt.show()
```

```python
# Хотите 6 - пожалуйста, также указываем расположение каждого axes и далее добавляем каждый axes на figur
fig = plt.figure(figsize=(8, 2)) # Чтоб места не занимать поставил маленький размер по высоте

ax_1 = fig.add_subplot(3, 2, 1) # Кол-во строк, кол-во столбцов, место в матрице графиков
ax_2 = fig.add_subplot(3, 2, 2)
ax_3 = fig.add_subplot(3, 2, 3)
ax_4 = fig.add_subplot(3, 2, 4)
ax_5 = fig.add_subplot(3, 2, 5)
ax_6 = fig.add_subplot(3, 2, 6)

ax_1.set(title = 'ax_1', xticks=[], yticks=[])
ax_2.set(title = 'ax_2', xticks=[], yticks=[])
ax_3.set(title = 'ax_3', xticks=[], yticks=[])
ax_4.set(title = 'ax_4', xticks=[], yticks=[])
ax_5.set(title = 'ax_5', xticks=[], yticks=[])
ax_6.set(title = 'ax_6', xticks=[], yticks=[])

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols =2 )

axes[0,0].set(title='axes[0,0]')
axes[0,1].set(title='axes[0,1]')
axes[1,0].set(title='axes[1,0]')
axes[1,1].set(title='axes[1,1]')

# axes.flat это метод итерации по массиву; np.ndarray.flat у нас по массиву axes!!!
for ax in axes.flat: # Цикл для установки одинаковых параметров для всех axes
    ax.set(xticks=[], yticks=[])
    
plt.show()
```

Заголовки можно часто установить в том же самом цикле: 

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

n = 1

for ax in axes.flat:
    ax.set(title='axes_' + str(n), xticks=[], yticks=[])
    n += 1
plt.show()
```
